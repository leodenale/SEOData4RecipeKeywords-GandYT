```python
import pytrends
from pytrends.request import TrendReq
import pandas as pd
import time
import datetime
from datetime import datetime, date, time
```

# youtube Trends

```python
import matplotlib.pyplot as plt
import seaborn as sns


#to get interest over time score, you'll need pytrend.interest_over_time() function.
#For more functions, check this: https://github.com/GeneralMills/pytrends

#
# Keyword List
#

pytrend = TrendReq()
pytrend.build_payload(kw_list=['H5n1','ASP', 'BSE'], timeframe='today 12-m', geo = 'DE', gprop='youtube')
interest_over_time_df1 = pytrend.interest_over_time() 

print(interest_over_time_df1.head())
#to get interest over time score, you'll need pytrend.interest_over_time() function.
#For more functions, check this: https://github.com/GeneralMills/pytrends

pytrend.build_payload(kw_list=['Hafer','Weizen', 'Getreide','Raps', 'Soja'], timeframe='today 12-m', geo = 'DE', gprop='youtube')
interest_over_time_df2 = pytrend.interest_over_time() 
print(interest_over_time_df2.head())

pytrend.build_payload(kw_list=['topagrar','proplanta', 'agrarheute','Bauernverband','f3'], timeframe='today 12-m', geo = 'DE', gprop='youtube')
interest_over_time_df3 = pytrend.interest_over_time() 
print(interest_over_time_df3.head())

pytrend.build_payload(kw_list=['Prognose','Vorhersage', 'Voraussage','Prophezeiung','Verkündung'], timeframe='today 12-m', geo = 'DE', gprop='youtube')
interest_over_time_df4 = pytrend.interest_over_time() 
print(interest_over_time_df4.head())



#
# Graphs
#

sns.set(color_codes=True)
dx = interest_over_time_df1.plot.line(figsize = (9,6), title = "Interest Over Time youtube")
dx.set_xlabel('Date')
dx.set_ylabel('Trends Index')
dx.tick_params(axis='both', which='major', labelsize=13)

sns.set(color_codes=True)
dx2 = interest_over_time_df2.plot.line(figsize = (9,6), title = "Interest Over Time youtube")
dx2.set_xlabel('Date')
dx2.set_ylabel('Trends Index')
dx2.tick_params(axis='both', which='major', labelsize=13)

sns.set(color_codes=True)
dx2 = interest_over_time_df3.plot.line(figsize = (9,6), title = "Interest Over Time youtube")
dx2.set_xlabel('Date')
dx2.set_ylabel('Trends Index')
dx2.tick_params(axis='both', which='major', labelsize=13)

sns.set(color_codes=True)
dx2 = interest_over_time_df4.plot.line(figsize = (9,6), title = "Interest Over Time youtube")
dx2.set_xlabel('Date')
dx2.set_ylabel('Trends Index')
dx2.tick_params(axis='both', which='major', labelsize=13)


```
